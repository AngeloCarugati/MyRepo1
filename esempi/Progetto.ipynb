```python
from gensim.models import Word2Vec
import numpy as np
import pandas as pd
from transformers import BertTokenizer
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize 
import string
from nltk.corpus import stopwords

with open('/home/acarugat/Testi/Presidenti/MattarellaFine2024.txt', 'r') as matta24:
    mattarella24 = matta24.read()
with open('/home/acarugat/Testi/Presidenti/MattarellaFine2023.txt', 'r') as matta23:
    mattarella23 = matta23.read()
    
sentences = sent_tokenize(mattarella24, language='italian')
print (sentences)
```

```python
import re

def preprocess_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
        for line in infile:
            # Dividi la riga in frasi
            sentences = re.split(r'[.!?]', line.strip())
            for sentence in sentences:
                if sentence:  # Ignora frasi vuote
                    outfile.write(sentence.strip() + '\n')  # Scrivi ogni frase su una nuova riga

# Pre-processa il file
preprocess_file("/home/acarugat/Testi/Presidenti/MattarellaFine2024.txt", "/home/acarugat/Testi/Presidenti/MattarellaFine2024-preprocessato.txt")
print("File preprocessato!")

# Usa il file preprocessato con Word2Vec
model = Word2Vec(corpus_file="/home/acarugat/Testi/Presidenti/MattarellaFine2024-preprocessato.txt", vector_size=100, window=5, min_count=1, workers=4)
model.save("word2vec_model.model")
print("Modello addestrato e salvato!")
```
